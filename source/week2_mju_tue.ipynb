## **1.2 문장(Statement)**

## **2.3. 여러 개의 값 출력**
여러개 값 각각은 콤마(,)로 구분, 공백 1 칸씩 띄어 출력

## 2.4. **출력 제어 문자**
\n : 새로운 줄로 이동, \t : 다음 탭으로 이동

````python
# 출력 제어 문자
print("\n줄바꿈\n연습")
print("\t탭키\t연습")
print("글자가\"강조\"되는 효과1")
print('글자가\'강조\'되는 효과2')
print("\\\\\\역슬래시 세 개 출력")
print("글자가 '강조'되는 효과")
````

## 2.5. **문자열 + : 병합**
문자+문자 = 문자병합, 숫자+숫자=연산

````python
# 문자열간의 +
print("100"+"200")
print(100+200)
# 문자 + 숫자
print("100" +200)
````

````python
#문제
print('이름:mju\n학번:60250001')
print("이름:mju\t학번:60250001")
````

# **3. 자료형(Data Type)**
- 정수형, 실수형, 문자열형, 논리형

## 3.1. **정수형(int)**
소수점 없는 숫자, 파이썬 정수는

```
# 코드로 형식 지정됨
```

크기 제한 없음

````python
# 정수형 연습
print(123)
print(123_456_789)
print(100**100)

````

## 3.2. **실수형(float)**
소수점 포함한 수, 부동소수점(3.14*e12 <-3.14 * 10<sup>10</sup>)

````python
# 실수형 연습
print(3.14)
print(3.14e-12)
2.0-1.9  # 이진으로 저장됨으로 오차가 생길 수 있음
````

## 3.3. **문자열형(str)**
단일 문자 집합, 작은따옴표/큰따옴표 로 감싼 문자 묶음

````python
# 문자열형 연습
print("Hello")
print('hello')
print("hello world") # 중간에 띄어쓰기도 괜찮음
````

### ▶ **문자열 + : 병합**
문자+문자 = 문자병합, 숫자+숫자=연산

````python
# 문자열 병합 연습
# 문자 + 문자 = 병합
print("100"+"200")
# 숫자 + 숫자 = 연산
print(100+200)
````

````python
# 문자 + 숫자 => 에러
print("100"+200)
````

````python
# 정수변환 이용
print(int("100") + 200)
````

## 3.4. **논리형(bool)**
참과 거짓으로 표현, True(참), False(거짓)

````python
# 논리형 연습
print(True)
print(False)
````

## 3.5. **자료형 확인 type()**
정수(int), 실수(float), 문자열(str), 논리(bool)

````python
# 자료형 확인 연습
print(type(123))
print(type(3.14))
print(type('python'))
print(type(True))



````

## 3.6. **자료형 정수 변환 int()**
문자열이나 실수를 정수로 강제 변환

````python
# 정수 변환 연습
print(int("12345"))
print(type(int("12345")))
print(int(3.141592))
print(int(1000))
````
