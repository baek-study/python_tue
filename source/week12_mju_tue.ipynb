##  **1.2 함수 호출 - 부르기**

*   수행할 함수의 이름을 써주는 것
*   함수이름(인수1, 인수2, .... 인수 n)
* 매개변수, 인수, 반환값, 결과 변수

```python
# get_area 함수 정의
# 매개변수 : radius,  결과(return) 있는 함수
# 하는 일 : radius*radius*3.14 를 계산 하여 리턴
def get_area(radius) :
    area = radius*radius*3.14
    return area

# print_msg 함수 정의
# 매개변수 : name, 결과(return) 없는 함수
# 하는일 : 안녕하세요, name님! 축하드립니다. 출력
def print_msg(name) :
    print('안녕하세요')
    print(f'{name}님! 축하드립니다.')

value = get_area(10)
print(f'원의 넓이 {value=}')
value = get_area(25)
print(f'원의 넓이 {value=}')

print_msg("명지")
```

## **1.3 지역 변수 vs 전역변수**


*   지역 변수 : 함수 내에서 생성된 변수와 매개변수
  - 선언된 함수 내에서만 접근 가능
  - 지역 변수 마다 동일한 이름 사용 가능 - 서로 다른 영역

* 전역 변수 : 함수 외부에서 생성되며 모든 함수에서 접근 가능
  - 함수내 전역 변수 변경 - 새로운 지역 변수로 간주
  - 함수내 global 키워드 이용시 - 전역 변수 변경 가능

```python
# 지역변수 -  선언된 함수 내에서만 접근 가능
# def myfunc():    - x : 3
def myfunc() :
    x = 3
    print(f'{x=}')
    myfunc2()
    print(f'{x=}')

# myfunc2() 함수  x : 10
def myfunc2() :
    x = 10
    print(f'{x=}')

# 주 프로그램
myfunc()
```

```python
# 전역변수 - 함수 외부에 생성 되어 모두 접근, gx, lx

# def myfunc() :
def myfunc() :
    lx = 5
    global gx
    gx = 5
    print(f'{lx=}')
    print(f'{gx=}')

gx = 3
myfunc()
print(f'{gx=}')
```

## **1.2 리스트 인덱싱 & 슬라이싱**

* 인덱스 : 0부터 양의 정수 배정, 리스트 끝부터는 음의 정수 가능
* 인덱싱: 리스트이름[인덱스] - 한 요소 접근
* 슬라이싱: 리스트이름[시작:끝:스텝] - 시작에서 끝-1까지 부분 리스트  


```python
# 리스트 생성
temps=[28, 31, 33, 35, 27, 26, 25]

# 리스트 인덱싱 - 인덱스 3값 출력, 인덱스 -1값 출력
print(f'{temps[3]=}')
print(f'{temps[-1]=}')

# 리스트 슬라이싱 start= 2, stop = 5
# 인덱스 2,3,4 값 출력
print(f'{temps[2:5]=}')
# 시작의 디폴트는 0
print(f'{temps[:5]=}')
# 끝의 디폴트는 리스트길이
print(f'{temps[2:]=}')
# 스텝의 디폴트는 1
print(f'{temps[2:7:2]=}')
#리스트의 역순
print(f'{temps[::-1]}')
```

## **1.3  리스트 복사**

*   얕은 복사: 복사리스트 = 원본리스트  - 주소만 복사
*   깊은 복사: 복사리스트 = list(원본리스트) or 원본리스트[:] - 요소 모두 복사




```python
# 얕은 복사 - 주소만 같음 - 같은 객체 가리킴
# temps, values
# 리스트 생성
temps=[28, 31, 33, 35, 27, 26, 25]

values = temps

# temps 출력
print(f'{temps=}')
# values 출력
print(f'{values=}')
# 한 리스트만 바뀌어도, 모두 바뀜
temps[0] = 0

print(f'## 얕은 복사에서 temps[0] = 0 수정 후')
# temps 출력
print(f'{temps=}')
# values 출력
print(f'{values=}')

print(id(temps))
print(id(values))
```

```python
#  깊은 복사 - 요소까지 복사 - 다른 것이 됨
# 리스트 생성
temps=[28, 31, 33, 35, 27, 26, 25]

# 깊은 복사 - values - list 함수 이용
values = list(temps)
# temps 출력
print(f'{temps=}')
# values 출력
print(f'{values=}')

temps[0] = 0
print(f'## 깊은 복사에서 temps[0] = 0 수정 후')
# temps 출력
print(f'{temps=}')
# values 출력
print(f'{values=}')
```

```python
#  깊은 복사 - 요소까지 복사 - 다른 것이 됨
# 리스트 생성
temps=[28, 31, 33, 35, 27, 26, 25]

# 깊은 복사 - values  - [:] 이용
values = temps[:]
# temps 출력
print(f'{temps=}')
# values 출력
print(f'{values=}')

temps[0] = 0
print(f'## 깊은 복사에서 temps[0] = 0 수정 후')
# temps 출력
print(f'{temps=}')
# values 출력
print(f'{values=}')
```

*   슬라이싱 수정 : 리스트이름[시작:끝] = [교체데이터]
* [주의] 리스트이름[인덱스] = [교체테이터] (리스트내의 리스트로 간주됨)



```python
# 리스트 요소 수정
nums = [11, 22, 33, 44, 55]

print(f'{nums=}')

# 인덱싱 이용한 수정  - nums의 첫번째를  -1로 수정
nums[0] = -1
print(f'{nums=}')
# nums 출력


nums = [11, 22, 33, 44, 55]
# 슬라이싱 이용한 수정 - nums[1:3] = [100, 200, 300]
nums[1:3] = [100,200,300]

# nums 출력
print(f'슬라이싱 후 : {nums=}')

nums = [11, 22, 33, 44, 55]
# 인덱싱에 리스트로 수정 - 리스트안에 리스트  - nums[1] = [100, 200, 300]
nums[1] = [100,200, 300]

# nums 출력
print(f'인덱싱에 리스트로 수정  : {nums=}')
```

## **1.5  리스트 연산(+, *, in/notin)**

*   리스트 연결 + : 리스트1 + 리스트2
* 리스트 반복 *: 리스트 * 정수 (리스트 요소를 정수번 반복)
* 존재여부 in/not in: 값 in 리스트

```python
nums1 = [11, 22, 33]
nums2 = [44, 55]

# 1. 리스트 연결 (+)
# nums3는 nums1, nums2 연결
nums3 = nums1 + nums2
# nums3 출력
print(f'{nums3=}')


# 2. 리스트 요소 반복(*)
# nums4는 리스트 nums1을 2번 반복
nums4 = nums1 * 2
# nums4 출력
print(f'{nums4=}')


# 3. 리스트내 값 존재 여부 (in, not in)
nums = [11, 22, 33, 44, 55]

# nums에 22가 있는지  - 출력
print(f'{22 in nums =}')
# nums에 44가 없는지 - 출력
print(f'{44 not in nums =}')
# nums에 66가 없는지 - 출력
print(f'{66 not in nums =}')
```

```python
temps = [28, 31, 33 , 35, 27, 26, 25]

# 리스트 객체 이용 - item 이용
for item in temps:
    print(item)

print()
# 리스트 인덱스 이용 - range() 함수 이용
for i in range(7) :
    print(temps[i])
```
