## **문제1**
정수 변수 age, 실수 변수 height를 ,
사용자에게  입력 받은 후, 각각 정수 변환 실수 변환을 한 후<br>
age가 18세가 아닌지 프린트 <br>
age가 30에서 49사이 인지 프린트 <br>
height 가 170.5이상인지 프린트
1. 변수선언
2. 사용자 입력
3. 비교 연산자 -age가 18세가 아닌지 프린트
4. 비교 연산자 -age가 30과 49 사이 인지
5. 비교 연산자 -height 가 170.5이상인지 프린트
<br>
나이 : (입력) <br>
키 : (입력) <br>



```python
# 나이 : (입력) -> 정수 변환
age=int(input('나이를 입력하십시오:'))
# 키 : (입력) -> 실수 변환
height=float(input('키를 입력하십시오:'))
# age가 18세가 아닌지 프린트
print(age!=18)
# age가 30이상 49이하 인지 프린트
print(30<=age and age<=49)
# height 가 170.5이상인지 프린트
print(height>=170.5)
```

* 사용자에게 초 단위의 시간을 입력받아, 분-초 형태로 출력 <br>
1분은 60초,(몫과 나머지를 여러번 이용)
---
변환하고자 하는 시간(초)? (사용자 입력) <br>
000 초는 00 분 00 초이다.

```python
SEC_PER_MINUTE = 60
# 사용자에게 초 단위의 시간(input_data)을 입력
input_data = int(input("변환하고자 하는 시간(초) : "))

# 분(minute)-초(second) 형태로 출력
minute = input_data // SEC_PER_MINUTE
second = input_data % SEC_PER_MINUTE

print(input_data, "초는", minute, "분 ", second, "초 ")
```

# **1부 관계/논리 연산자**



## **1.1 관계/비교연산자**
두 개의 자료 또는 값(피연산자)을 비교하는 연산자<br>
결과값은 항상 참(true) 아니면 거짓(false)

```python
# n1에 10, n2에 100 저장
n1=10
n2=100
# n1과 n2가 같은지
print(n1==n2)
# n1과 n2가 같지 않은지
print(n1!=n2)
# n1과 n2보다 작은지
print(n1<n2)
# n1과 n2보다 큰지
print(n1>n2)
# n1과 n2보다 작거나 같은지
print(n1<=n2)
# n1과 n2보다 크거나 같은지
print(n1>=n2)
```

### 관계/비교 연산자 주의 사항 ###


1.   == : 같다, = : 대입 연산자
2.   논리형 비교 : ==, !=

```python
# 1. == : 같다, = : 대입 연산자
n1 = 3
n2 = 3
# 같다는 표현
print(n1==n2)
# 대입시는 ?
#print(n1=n2)
```

```python
# 2. 논리형 비교 : ==, !=
# exp에 참 대입
exp = True
# exp가 참 인지
print(exp == True)
# exp가 거짓 인지
print(exp != False)
```

```python
# 3. 실수시 비교 : 근사값 저장됨
import math
n = math.sqrt(3)   # 루트 값
# n의 제곱이 3인지 출력
print(n*n == 3.0)
# 결과의 이유?
print(math.fabs(n*n- 3.0)<0.000001)
```

### **if구문 여러가지 버전**


*   한줄 버전 : if 조건: 문장
*   기본 버전<br>
if 조건: <br>
[빈칸]문장
* 여러문장 제어 버전<br>
if 조건: <br>
[빈칸]문장1<br>
[빈칸]문장2



```python
score = int(input("점수는 ? "))

# 만약 score 가 60 이상 이면 "pass" 프린트 - 조건
# 한줄 버전
if score>=60: print("pass")

# 기본 버전
if score>=60 :
    print("pass")
# 여러문장 제어 - "pass" 출력, 다음문장은 "score는 00 입니다." 출력

if score>=60:
    print("pass")
    print("패스")

if score>=60:
    print("pass")
print("패스")
```

## **3.2 f-문자열:자료형식**
{표현식:자료형식코드}

*   정수 : d(decimal), {num:d}
*   실수 : f(fixed/float), {avg:f}
* 문자열 : s (string), {msg:s}



```python
num = 15

print(f'num = [{num:5d}]')
print(f'pi = {3.14566:.2f}')

# num 출력 - 10진 표현

n1 = 1000
n2 = 8

avg = n1/n2

# avg 출력
print(f'{avg}')
# avg 출력 - 실수 f 형식
print(f'실수 f 형식 {avg:f}')
# avg 출력 - 실수 e 형식
print(f'실수 e 형식 {avg:e}')
```

### **자릿수 & 정렬**


*   {num:5d}   # 5자리 확보
*   {num:<5d}  # 왼쪽 정렬
*   {num:<05d} # 빈공간에 0



```python
num = 15

# 다섯자리 확보
print(f'다섯자리 확보 {num:5d}')
# 왼쪽 정렬
print(f'왼쪽 정렬 {num:<5d}')
# 5자리면서 빈공간에 0 채우기
print(f'빈공간에 0 채우기 {num:05d}')
```

### **정밀도:소수점 몇자리**


*  {PI:f}# 기본-소수점 이하 6자리
*   {PI:.2f} # 소수점 이하 두자리
* {PI:5.2f} # 전체가 5자리(소수점 포함)



```python
PI = 3.14159265358979
print(f'기본-소수점 이하 6자리 {PI:f}')
print(f'소수점 이하 두자리 {PI:.2f}')
print(f'전체가 5자리(소수점 포함)[{PI:5.2f}]')
```
