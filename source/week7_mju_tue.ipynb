## **문제1**
더하기/나누기만 가능한 계산기를 만들고자 한다.  
- 사용자에게 수행할 연산을 입력 받은 후, 연산 결과를 result에 대입.
- num1, num2는 1에서 45까지 임의의 값  (이때, import random 되있다고 가정) <br>
1) num1, num2 - 임의의 값 <br>
2) op 변수에 수행할 연산을 사용자에게 입력 ( +, /)<br>
3) 만약 op 가 '+' 이면 - result = ?<br>
4) 그렇지 않고 op 가 '/' 이면 - result = ?<br>
5) 그렇지 않으면 (op가 +, / 도 아니면) result = 0 .<br>






```python
import random
# 1) 변수 할당 num1, num2 - 임의의 값
num1 = random.randint(1,45)
num2 = random.randint(1,45)
# 2) op 변수에 수행할 연산을 사용자에게 입력 ( +, /)
op = random.choice(['+', '/'])
# 3) 만약 op 가 '+' 이면 - result = ?
if op == '+' :
    print(f'result = {num1+num2}')
    print(f'{num1+num2=}')
# 4) 그렇지 않고 op 가 '/' 이면 - result = ?
elif op == '/' :
    print(f'result = {num1/num2}')
    print(f'{num1/num2=}')
# 5) 그렇지 않으면 (op가 +, / 도 아니면) result = 0
else :
    print(f'result = 0')
```

## **2.2 문자열 접근**
* 인덱싱 : 개별 문자
* 슬라이싱: 부분 문자열

## 1. 인덱싱 : [index]
- 인덱스는 0부터 시작, 특정위치 반환


```python
## 1. 인덱싱 : [index] 특정위치 반환 - 0 부터 시작
msg = '안녕하세요'
print(msg[0])
print(msg[1])
print(msg[2])
print(msg[3])
print(msg[4])
#print(msg[5])
print(msg[-1])
print(msg[-2])
# 0 번째 인덱스
print(f'{msg[0]}')
# -1번째 인덱스
```

## 2. 슬라이싱 : [start:end]
* 시작부터 끝 직전(끝-1) 위치까지 부분 문자열 반환

```python
## 2. 슬라이싱 : [start:end:step] 시작부터 끝 직전(끝-1) 위치까지 부분 문자열 반환
msg = '안녕하세요'
# 문자열 인덱스 0 부터 2까지
print(msg[0:2])  # msg[0], msg[1]
# 문자열 인덱스 2부터 5까지
print(msg[2:5])  #[2], [3], [4]
# 문자열 시작 생략 가능 : 문자열 첫 문자부터
print(msg[:2])
# 문자열 끝 생략 : 문자열 마지막 문자까지
print(msg[2:])
```

```python
## 2. 슬라이싱 : 문자열을 2부분으로 나누기

msg = "20250425Rainy"

# date
date = msg[:8]
print(date)
# weather
print(weather := msg[8:])

# year
print(year := date[:4])
# month
print(mon := date[4:6])
# day
print(day := date[6:8])
```

## **2.3 문자열 연산**
* \+ : 두 문자열 병합
* \* : 문자 반복
* in, not in : 문자열 내 반

```python
## 3. 문자열 연산 +
msg1, msg2 = '안녕', '하세요'  # 다중 할당문

# + : 문자간 연결 - msg1, msg2 연결
print(msg1 + msg2)
print(msg1, msg2)
```

```python
## 3. 문자열 연산 *  : 반복
msg1 = '안녕'
# msg1 3번 반복
print(msg1*3)
print(3*msg1)

#[강의자료 예제]
# '#' 다섯번 반복
print('#'*5)
print(5*'#')
```

```python
## 3. 문자열 연산 in, not in :
msg = '안녕하세요'
#'안' 문자가 msg에 포함되었는지
print('안' in msg)
# not in : '안' 문자가 msg에 포함되지 않았는지
print('안' not in msg)

#[강의자료 예제]
text = 'python'

#'t'가 text에 있는지
print('t' in text)
#'T'가 text에 있는지
print('T' in text)
#'th'가 text에 있는지
print('th' in text)
#'티'가 text에 있지 않은지
print('티' not in text)
```
