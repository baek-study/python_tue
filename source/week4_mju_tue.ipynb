### **변수 할당/대입 - 자료/수식/함수**
오른쪽의 값(결과값)이 왼쪽의 변수 공간에 저장 <br>

* 사용자에게 초 단위의 시간을 입력받안 시-분-초 형태로 출력 <br>
1분은 60초, 1시간, 3600초
---
변환하고자 하는 시간(초)? (사용자 입력) <br>
000 초는 00 시간 00 분 00 초이다.



```python
# 기호 상수
HOUR_TO_SECOND = 3600
MINUTE_T0_SECOND = 60
#  사용자에게 초 단위의 시간을 입력받안 시-분-초 형태로 출력


x = int(input("초를 입력하시오"))
hour = x//HOUR_TO_SECOND
temp = x%HOUR_TO_SECOND
minute = temp//MINUTE_T0_SECOND
second = temp%MINUTE_T0_SECOND
print(x, "초는", hour, "시간", minute, "분", second, "초")
```

## **2.2 대입연산자**
우측 표현식의 결과값을 좌측 변수에 대입 <br>
다중 할당문<br>
- 한번에 여러개 변수 초기화 <br>
- 두 변수 맞교환

```python
# 다중 할당문 - 여러 개의 대입 연산
x = y = z =1
print(x, y, z)

# 좌변 변수가 여러개 인 경우 - 다중 대입
x, y = 2, 3
print(x, y)

x, y, z = 2,5, 7
print(x, y, z)


# 두 변수 맞교환
x, y = y, x
print(x, y)

# 다른 언어에서 맞교환
temp = x
x = y
y = temp
print(x, y)
```

## **2.3 복합대입연산자**
+=처럼 대입 연산자와 다른 연산자를 합쳐 놓은 연산자 <br>

```python
num = 10
# 복합대입 num = num + 2
num += 2
print(num)
# 복합대입 num = num - 2
num -= 2
print(num)
# 복합대입 num = num * 2
num *= 2
print(num)
# 복합대입 num = num / 2
num /= 2
print(num)
# %
num %= 2
print(num)
```

## **2.5 비교연산자**

```python
# n1에 2, n2에 4 저장
n1, n2 = 2, 4
# n1과 n2가 같은지
print(n1==n2)
# n1과 n2가 같지 않은지
print(n1!=n2)
# n1과 n2보다 큰지
print(n1>n2)
# n1과 n2보다 크거나 같은지
print(n1>=n2)
# n1과 n2보다 작은지
print(n1<n2)
# n1과 n2보다 작거나 같은지
print(n1<=n2)
```
