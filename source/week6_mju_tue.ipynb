## **2.2 연속 if : if ~ elif ~ else**

```python
# 사용자가 입력한 점수(score)가 90 이상이면 A, 그렇지 않고 80이상이면 B, 그렇지 않고 70 이상이면 C, 그렇지 않고 60 이상이면 D, 그렇지 않으면 F
print('#'*5, 'if ~ else if ~ else 문', '#'*5)
score = int(input("점수를 입력하시오:"))

# 만약 score가 90 이상이면 - A
if score >= 90 :
    print("A학점입니다.")
# 그렇지 않고 만약 score가 80 이상이면 - B
elif score >= 80 :
    print("B학점입니다.")
# 그렇지 않고 만약 score가 70 이상이면 - C
elif score >= 70 :
    print("C학점입니다.")
# 그렇지 않고 만약 score가 60 이상이면 - D
elif score >= 60 :
    print("D학점입니다.")
# 그렇지 않으면 - F
else :
    print("F학점입니다.")

print('end!')
```

### 가위, 바위, 보 게임을 통해서 컴퓨터와 대결해 봅시다.
* 잘 알겠지만, 가위바위보 게임은 다음의 규칙을 따릅니다. 나를 기준으로 컴퓨터가 내는 것으로 확인하면 됩니다.

```python
# my_hand vs com_hand

import random

# 사용자의 가위/바위/보 중 하나를 입력받아 myHand 변수에 저장
my_hand = input("가위/바위/보를 선택하세요 : ")

# 컴퓨터의 가위/바위/보는 랜덤하게 추출하여 comHand에 저장
com_hand = random.choice(['가위', '바위', '보'])

print(f'my_hand는 {my_hand}, com_hand는 {com_hand}')
print(f'{my_hand=}, {com_hand=}')
# 게임 진행하기 위해 사용자의 가위/바위/보에 따라 if문을 사용함
# my_hand가 가위 이면 - 가위라고 출력
if my_hand == '가위':
    print('my_hand는 가위입니다')
# my_hand가 바위 이면 - 바위라고 출력
elif my_hand == '바위':
    print('my_hand는 바위입니다' )
# my_hand가 보 이면 - 보라고 출력
elif my_hand == '보':
    print('my_hand는 보입니다' )
else :
    print('이외는 잘못된 것입니다.')
```

```python
# my_hand vs com_hand
import random

# 사용자의 가위/바위/보 중 하나를 입력받아 myHand 변수에 저장
my_hand = input("가위/바위/보 중 하나를 선택하시오 : ")

# 컴퓨터의 가위/바위/보는 랜덤하게 추출하여 comHand에 저장
com_hand = random.choice(["가위", "바위", "보"])

print("my_hand = ", my_hand, "com_hand =", com_hand)
# my_hand가 가위 이면 -
if my_hand == "가위" :
    ## com_hand가 가위 이면 비김
    if com_hand == "가위" :
        print("비겼습니다.")
    ## com_hand가 바위 이면 "졌습니다" 출력
    elif com_hand == "바위" :
        print("졌습니다")
    ## com_hand가 보 이면 "이겼습니다" 출력
    else :
        print("이겼습니다.")
# my_hand가 바위 이면 - 바위라고 출력
elif my_hand == "바위" :
    ## com_hand가 가위 이면 비김
    if com_hand == "가위" :
        print("이겼습니다.")
    ## com_hand가 바위 이면 "졌습니다" 출력
    elif com_hand == "바위" :
        print("비겼습니다")
    ## com_hand가 보 이면 "이겼습니다" 출력
    else :
        print("졌습니다")
# my_hand가 보 이면 - 보라고 출력
elif my_hand == "보" :
    ## com_hand가 가위 이면 비김
    if com_hand == "가위" :
        print("졌습니다")
    ## com_hand가 바위 이면 "졌습니다" 출력
    elif com_hand == "바위" :
        print("이겼습니다")
    ## com_hand가 보 이면 "이겼습니다" 출력
    else :
        print("비겼습니다")
else :
    print("가위/바위/보 만 낼 수 있습니다.")
```

### 실습문제 ###
키와 몸무게를 입력하면 키에 맞는 표준 체중을 구한 후, 자신의 몸무게와 표준 체중과의 차이를 기준으로 저체중, 표준, 과체중 중에 하나를 출력하는 프로그램 작성


*   bmi(표준 체중) = (자신의 키 - 100) * 0.9
* '표준’: (표준체중 – 3kg) 이상이면서  (표준체중 + 3kg) 이하인 체중
* '저체중’: (표준체중 – 3kg)보다 작은 체중
*‘과체중’: (표준체중 + 3kg)보다 큰 체중





```python
# 키(height) 사용자 입력
height = float(input("키를 입력하시오"))
# 몸무게(weight) 사용자 입력
weight = float(input("몸무게를 입력하시오"))
# bmi 계산
bmi = (height-100)*0.9
# 연속 if
if weight>=(bmi-3) and weight<=(bmi+3):
    print('표준 입니다.')
elif weight > (bmi+3) :
    print('과체중입니다.')
else:
    print('저체중입니다.')
```

### 실습 문제
입력한 연도가 윤년인지 아닌지를 판단하여 화면에 출력하는 프로그램을 작성


*  율리우스력 : 입력 년도를 4로 나눈 나머지가 0이면 윤년
*  그레고리력 <br>
입력 년도를 4로 나눈 나머지가 0 이고,
     100으로 나눈 나머지가 0이 아니면 윤년<br>
400 으로 나누어 떨어지면 윤년





```python
# year 사용자 입력 - 년도를 입력하시오
year=int(input('년도를 입력하시오'))
# 율리우스력
# 만약 년도가 4로 나눈 나머지가 0 이면
if year%4==0:
    print(year,'은 율리우스력으로 윤년입니다.')
else:
    print(year,'은 율리우스력으로 윤년이 아닙니다.')

# 그레고리력 - 먼저 조건식 먼저
# (1) 년도가 4로 나눈 나머지가 0 이고, 100으로 나눈 나머지가 0이 아닌 경우
# year%4==0 and year%100!=0
# (2) 년도가 400으로 나눈 나머지가 0인 경우 윤년
#year%400 == 0

if (year%4==0 and year%100!=0) or (year%400 == 0 ) :
    print(year,'은 그레고리력으로 윤년입니다.')
else :
    print( year, '은 그레고리력으로 윤년이 아닙니다.')
```
