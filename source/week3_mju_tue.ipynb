##**2.2 입력함수 + 타입 변환**
- 정수로 변환 : int(input())<br>
- 실수로 변환 :  float(input())

```python
# 방법 1 : input() 함수로 변수에 입력 저장 -> 변수를 타입 변환
# '정수를 입력하시오'를 화면에 출력후 사용자 입력을 받은 후 변수 n1에 저장
n1 = input("정수를 입력하시오 :")
print(type(n1))
# n1을 정수로 변환
n1 = int(n1)
print(type(n1))
```

```python
# 방법 2 : input() 함수 결과를 바로 int로 변환 후 바로 변수에 저장
# '정수를 입력하시오'를 화면에 출력후 사용자 입력을 받고 바로 정수 변환 함수 호출 -> 변수 n1에 대입
n1 = int(input("정수를 입력하시오:"))
print(type(n1))
```

```python
# 국어 점수를 입력 -> 정수 변환 -> 국어 점수 출력 (방법1 사용)
kor_score_1 = input("국어 점수를 입력하시오(방법1):")
kor_score_1 = int(kor_score_1)
print(kor_score_1)
# 국어 점수를 입력 -> 정수 변환 -> 국어 점수 출력 (방법2 사용)
kor_score_2 = int(input("국어 점수를 입력하시오(방법2):"))
print(kor_score_2)
# 국어 점수, 수학 점수를 더해서 sum에 저장
sum = kor_score_1 + kor_score_2
print(sum)
```
