###  **함수 호출 - 부르기**


*   수행할 함수의 이름을 써주는 것
*   함수이름(인수1, 인수2, .... 인수 n)

```python
# get_area 함수
# 매개변수 : radius,  결과(return) 있는 함수
# 하는 일 : radius*radius*3.14 를 계산 하여 리턴
def get_area(radius) :
    area = 3.14*radius**2
    return area

#호출
value1 = get_area(3)
value2 = get_area(10)

print(f'반지름 3의 원의 넓이 : {value1=},반지름 10의 원의 넓이 : {value2=}')
# print_msg 함수
# 매개변수 : name, 결과(return) 없는 함수
# 하는일 : 안녕하세요, name님! 축하드립니다. 출력
def print_msg(name) :
    print("안녕하세요?")
    print(name+"님! 축하드립니다.")

#호출
print_msg("영희")
print_msg("명지")


# make_msg 함수
# 매개변수 : 없음, 결과(return) 없는 함수
# 하는일 :  사용자에게 name 입력 받은 후, 안녕하세요, name님! 축하드립니다. 화면에 출력
def make_msg() :
    name =input("이름은?")
    msg ="반가워요"+name
    print(f'{msg}')

#호출
make_msg()
make_msg()

# make_msg_return 함수
# 매개변수 : 없음, 결과(return) 있는 함수
# 하는일 :  사용자에게 name 입력 받은 후, msg에 "안녕하세요, name님! 축하드립니다."를 넣은 후 리턴
def make_msg_return() :
    name =input("이름은?")
    msg ="반가워요"+name
    return msg

#호출
value4 = make_msg_return()
value5 = make_msg_return()

print(f'{value4=}, {value5=}')
```

##  **인수 종류**
*   위치 인수 : 인수를 순서대로 값을 전달
*   키워드 인수 : 이름  - 값
* 기본값(디폴트) 인수  : 매개변수에 기본값 정의
* 가변 인수 : 인수 개수가 정해지지 않음


```python
# 매개변수 x, y
# 함수이름 get_max
# 리턴값 있음
# 하는 일은 : x가 크면  x를 리턴 y 가 크면 y 리턴
def get_max(x,y=10):
     if x>y:
        return x
     else:
        return y

# 호출 - value에 ...
value= get_max(4,5)
print(f'{value=}')
value= get_max(4)
print(f'{value=}')

value= get_max(x=4,y=5)
print(f'{value=}')
value= get_max(y=5,x=4)
print(f'{value=}')
```

```python
# get_area
def get_area(radius=1) :
    area = 3.14*radius**2
    return area

# 위치 인수
value = get_area(3)
# 키워드 인수
value = get_area(radius=3)
# 기본값(디폴트) 인수
value = get_area()

# 가변 인수
def sub(*x) :
    print(x)

sub("mju", "아프다", "우린 웃기다")
```

# **2부 지역변수와 전역변수**
지역변수(local variable), 전역변수(global variable)

## **2.1 지역 변수**


*   함수 내에서 생성된 변수와 매개변수
*   선언된 함수 내에서만 접근 가능



```python
# 지역변수 1/2 -  선언된 함수 내에서만 접근 가능
# def myfunc():
def myfunc() :
    x = 3
    print(f'myfunc에 있는 {x=}')
    myfunc2()
    print(f'myfunc에 있는 {x=}')

def myfunc2() :
    x = 100
    print(f'myfunc2에 있는 {x=}')

# 주 프로그램
myfunc()

# 함수 외에서 지역 변수 접근 불가능
print(f'바깥쪽에서 {x=}')
```

```python
# 지역변수 2/2 - 지역 변수 마다 동일한 이름 사용 가능
# def myfunc():
def myfunc():
    x = 3
    myfunc2()
    print(x)

# myfunc2() 함수
def myfunc2():
    x = 100
    print(x)

# 주 프로그램
myfunc()
```

## **2.2 전역 변수**
함수 외부에서 생성되며 모든 함수에서 접근 가능

```python
# 전역변수 1/2 - 함수 외부에 생성 되어 모두 접근
gx = 3         # 전역변수

def myfunc() :
    lx = 5      # 지역변수
    global gx
    gx = 5
    print(f'{lx=}')
    print(f'{gx=}')

myfunc()
print(f'{gx=}')
```

```python
# 함수 내부에서 전역 변수 사용 가능
# get_circle_area   - 전역 변수 r  사용 가능
# 매개변수 radius, 반환 : 없음
# 하는 일  area 에 원의 넓이 구하는 것.

def get_circle_area(radius):
    global area
    area=radius**2*3.14

# 전역 변수, r, area 가 있을때...

r= float(input("반지름을 입력하시오: "))
area=0
get_circle_area(r)
print(f"반지름 {r}의 넓이는 {area}")
```

## **2.3  전역변수 변경- global 키워드 이용**


1. 함수내 전역 변수 변경 - 새로운 지역 변수로 간주
2. 함수내 global 키워드 이용시 - 전역 변수 변경 가능  



```python
# 1. 함수 내에서 전역 변수 변경 - 새로운 지역 변수 생성
gx = 3

def myfunc():
    lx = 5
    gx = 10                #  새로운 지역 변수로
    print(lx)
    print(gx)

myfunc()
print(gx)
```

```python
# 2. 함수 내에서 전역 변수 변경 - global 키워드 이용시
gx = 3

def myfunc():
    lx = 5
    global gx
    gx = 10                #  전역변수 사용
    print(lx)
    print(gx)

myfunc()
print(gx)
```

##**3.1 모듈 활용**
특정모듈을 현재 모듈로 가져오는 5가지 방법



```python
## 방법1 - import 모듈이름
import random

number = random.randint(1, 45)
print(f'로또 복권 : {number}')

## 방법2 - import 모듈이름 as 별명   - alias
import random as rd
number = rd.randint(1, 45)
print(f'로또 복권 : {number}')
```

```python
## 방법3 - from 모듈이름 import 함수이름, 함수이름
from random import randint

number = randint(1, 8)
print(f'로또 복권 : {number}')


## 방법4 - form 모듈이름 import 함수 이름 as 별명
from random import randint as rdi
number = rdi(1, 8)
print(f'로또 복권 : {number}')
```

```python
## 방법5 - form 모듈이름 import *  - 모듈내 모든 함수 가져옴
from random import *
number = randint(1, 8)
print(f'로또 복권 : {number}')
```

```python
%%writefile cal_area.py
# 파일명 : calc_area.py
# 설  명 : 도형의 넓이를 구하는 모듈

# 1. 원의 넓이를 구하기 get_circle_area 함수
# 매개변수 : 반지름, 반환값 : 원의 넓이
def get_circle_area(radius):
    result = 3.14*radius**2
    return result

# 2. 사각형의 넓이를 구하기
# 매개변수 : 너비, 높이, 반환값 : 사각형의 넓이
def get_rect_area(width, height) :
    result = width*height
    return result
```
