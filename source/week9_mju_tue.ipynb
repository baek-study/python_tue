### **시험문제1**
더하기/나누기만 가능한 계산기를 사용자에게 정수(num)을 입력 받아, 각 자릿수의 합을 출력하는 프로그램을 작성 하시오(4). <br>
   (while 구문 이용하기)<br>
   ex) 123 => 1 + 2 + 3 = 6 <br>
“123의 자릿수의 합은 [   6]입니다.”
이때, f-문자열로 출력하고, 정수는 4자리 <br>





```python
# while
num = int(input("정수 입력"))
sum = 0   # 합을 위한 변수 - 초기화
temp = num

while temp > 0 :
    sum = sum + temp % 10
    temp = temp // 10

print(f'{num}의 자릿수의 합은 {sum:4d}')
```

# **1부 문자열 관련 함수 - 메소드**

*   내장함수 - len(s), max(s), min(s)
*   문자열 메소드 - s.split(), s.count(), s.find()

## **1.1 내장 함수**
* len(s) : 길이
* max(s) : 가장 큰 값
* min(s) : 가장 작은

```python
msg = '파이썬 프로그램은 재밌다'

# 길이 함수
length = len(msg)
print(f'{msg}는 {length}자리 정수 입니다.')
# 가장 큰 값 구하는 함수
print(f'{msg}의 가장 큰 값은  {max(msg)} 입니다.')
# 가장 작은 값 구하는 함수
print(f'{msg}의 가장 작은 값은  {min(msg)} 입니다.')
```

```python
#[강의자료 예제] len()을 이용한 문자열 길이 확인 예
# 1. num에 사용자에게 정수 입력 - 정수를 입력하세요
num=input("정수 입력:")
# 2. len함수 이용해 길이 지정
length=len(num)
# 3. 길이 출력
print(length)
```

## **1.2 메소드**
* s.split() : 문자열 분리
* s.count('단어') : 문자열내 단어 갯수
* s.find('단어') : 문자열내 단어 위치

```python
msg = 'The python programming is fun'

# 문자열 분리
w1, w2, w3, w4, w5  = msg.split()
print(f'{w1=}, {w2=}, {w3=}, {w4=}, {w5=}')

# 문자열내 단어 카운트 - 'p', 'px'
print(msg.count('p'))
print(msg.count('px'))

# 문자열내 단어 찾기 -  'p', 'px'
print(msg.find('p'))
print(msg.find('px'))

# 문자열내 단어 바꾸기  replace 'python' - 파이썬
print(msg.replace('python', '파이썬'))
```

### **함수 정의 - 만들기**


*   def 함수이름(매개변수) :<br>
&emsp; 문장들 <br>
&emsp; return 값

```python
# get_area 함수
# 매개변수 : radius,  결과(return) 있는 함수
# 하는 일 : radius*radius*3.14 를 계산 하여 리턴
def get_area(radius) :
    area = radius*radius*3.14
    return area

# print_msg 함수
# 매개변수 : name, 결과(return) 없는 함수
# 하는일 : 안녕하세요, name님! 축하드립니다. 출력
def print_msg(name) :
    print("안녕하세요")
    print(f'{name}님!  축하드립니다 ')
```

###  **함수 호출 - 부르기**


*   수행할 함수의 이름을 써주는 것
*   함수이름(인수1, 인수2, .... 인수 n)


```python
# get_area 부르기
# radius  - 사용자에게 반지름 받기
print(get_area(3))
print(get_area(10))

# print_msg 부르기
print_msg("영희")
print_msg("철수")
```

 # **3부 매개변수/인수/반환값**


*   매개변수(parameter) : 함수 정의에서 인수를 전달받는 변수
*  인수(argument) : 함수 호출시 함수에게 전달되는 값
*   반환 값(return value) : 함수 수행의 결과로 호출자에게 돌려주는 값



```python
# 함수 정의 - get_area
def get_area(radius) :
    area = radius*radius*3.14
    return area

# 함수 호출 - get_area
value = get_area(3)

print(f'반지름 3의 원의 넓이 : {value=}')
```

```python
# 함수 정의
# get_sum 함수
# 매개변수 : first, second,third  결과(return) 있는 함수
# 하는 일 : 세 값을 더하는 함수

def get_sum(first,second,third):
    sum=first+second+third
    return sum

# 함수 호출 - get_sum
result=get_sum(1,2,3)
print(f'get_sum(1,2,3) 결과 : {result=}')
```
